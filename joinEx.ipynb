# Spark join types

```
tenure_data = [
    (201, 'Krishnan', 3), 
    (301, 'Murugan', 2),
    (401, 'Devi', 4),
    (501, 'Gauri', 1)
]

empl_data = [
    (101, 'Raman', 'Finance'), 
    (201, 'Krishnan',  'Finance'),
    (301, 'Murugan',  'HR'),
    (401, 'Devi',  'Sales')
]

employees = spark.createDataFrame(empl_data, ['id', 'name', 'department'])
tenure = spark.createDataFrame(tenure_data, ['id', 'name', 'tenure_yrs'])
employees.registerTempTable("emplTab")
tenure.registerTempTable("tenureTab")
```

```
tenure.show()
employees.show()
```

# Inner join vs Left join vs Right join vs Full join

```
tenure.join(employees, on='id', how='inner').show()
```

```
joinExpression = [tenure.id == employees.id]
tenure.join(employees, joinExpression, 'inner').show()
```

```
tenure.join(employees, on='id', how='inner').drop(tenure.name).show()
```

```
# inner join
spark.sql("""SELECT emplTab.id, emplTab.name, department, tenure_yrs
    FROM  tenureTab INNER JOIN emplTab 
    ON  tenureTab.id = emplTab.id""").show()
```

```

spark.sql("""SELECT emplTab.id, emplTab.name, department, tenure_yrs
    FROM emplTab NATURAL JOIN tenureTab """).show()
```

```
tenure.join(employees,  on='id', how='left').show()
```

```

spark.sql("""SELECT emplTab.id, emplTab.name, department, tenure_yrs
    FROM tenureTab LEFT JOIN emplTab
    ON emplTab.id = tenureTab.id""").show()
```

```
tenure.join(employees,  on='id', how='right').show()
```

```

spark.sql("""SELECT emplTab.id, emplTab.name, department, tenure_yrs
    FROM  tenureTab RIGHT  JOIN emplTab
    ON emplTab.id = tenureTab.id""").show()
```

```
tenure.join(employees, on='id', how='full').show()
```

```
# inner join
spark.sql("""SELECT *
    FROM  tenureTab FULL JOIN emplTab
    ON tenureTab.id = emplTab.id""").show()
```

# Full join  vs full outer join vs cross join

```
tenure.join(employees, on='id', how='full').show()
tenure.join(employees, on='id', how='fullouter').show()
```

```
tenure.crossJoin(employees).show()
```

```
# inner join
spark.sql("""SELECT *
    FROM  tenureTab CROSS JOIN emplTab""").show()
```

#Left join vs Left outer join vs Left anti join vs Left semi join

```
tenure.join(employees, on='name', how='left').show()
tenure.join(employees, on='name', how='left_outer').show()
```

```
tenure.join(employees, on='id', how='leftanti').show()
```

```

spark.sql("""SELECT * 
    FROM tenureTab   ANTI JOIN emplTab 
    ON tenureTab.id  = emplTab.id """).show()
```

```
tenure.join(employees, on='id', how='leftsemi').show()
```

```

spark.sql("""SELECT * 
    FROM tenureTab   SEMI JOIN emplTab 
    ON tenureTab.id  = emplTab.id """).show()
```

# Non equi joins

```
cond = [tenure.id == employees.id, employees.department == 'Finance']
tenure.join(employees, cond, 'inner').show()
```

```

spark.sql("""SELECT * 
    FROM tenureTab   INNER JOIN emplTab 
    ON tenureTab.id  = emplTab.id 
    where  emplTab.department == 'Finance'""").show()
```
